```python
# Import Package:
#---------------------
import pandas as pd
import numpy as np 
from datetime import datetime as dt
import time
import os 
```

```python
# Identify dirction and present Content of path::
#----------------------------
path = os.path.abspath('') 
print(f"The Path Dirction :: {path}\n----------------\n\nThe Content Path\n--------------------")
files = os.listdir(path)
for file in files :
    print(file)
```

```python
# Clear DF and looping over all files and Concat all files endwith(CSV) into one File::
#-------------------------------------------------------------------------------------------
df = pd.DataFrame()
for file in files:
    if file.endswith('.csv'):
        df = df.append(pd.read_csv(file), ignore_index=True) 
df.head() 
df.to_csv('All-data-ofAuto-after-Handle.csv' , index = False)
```
